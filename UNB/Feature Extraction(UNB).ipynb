### **3.1. Address Bar Based Features:**

Many features can be extracted that can be consided as address bar base features. Out of them, below mentioned were considered for this project.


*   Domain of URL
*   IP Address in URL
*   "@" Symbol in URL
*   Length of URL
*   Depth of URL
*   Redirection "//" in URL
*   "http/https" in Domain name
*   Using URL Shortening Services “TinyURL”
*   Prefix or Suffix "-" in Domain

Each of these features are explained and the coded below:

```
from urllib.parse import urlparse,urlencode
import ipaddress
import re
```

```
# 1.Domain of the URL (Domain) 
def getDomain(url):  
  domain = urlparse(url).netloc
  if re.match(r"^www.",domain):
	       domain = domain.replace("www.","")
  return domain
```

```
# 2.Checks for IP address in URL (Have_IP)
def havingIP(url):
  try:
    ipaddress.ip_address(url)
    ip = 1
  except:
    ip = 0
  return ip
```

```
# 3.Checks the presence of @ in URL (Have_At)
def haveAtSign(url):
  if "@" in url:
    at = 1    
  else:
    at = 0    
  return at
```

```
# 4.Finding the length of URL and categorizing (URL_Length)
def getLength(url):
  if len(url) < 54:
    length = 0            
  else:
    length = 1            
  return length
```

```
# 5.Gives number of '/' in URL (URL_Depth)
def getDepth(url):
  s = urlparse(url).path.split('/')
  depth = 0
  for j in range(len(s)):
    if len(s[j]) != 0:
      depth = depth+1
  return depth
```

```
# 6.Checking for redirection '//' in the url (Redirection)
def redirection(url):
  pos = url.rfind('//')
  if pos > 6:
    if pos > 7:
      return 1
    else:
      return 0
  else:
    return 0
```

```
# 7.Existence of “HTTPS” Token in the Domain Part of the URL (https_Domain)
def httpDomain(url):
  domain = urlparse(url).netloc
  if 'https' in url.lower():
    return 1
  elif 'http' in url.lower():
    return 0
```

```
# listing shortening services
shortening_services = r"bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|" \
                      r"yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|" \
                      r"short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|" \
                      r"doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|db\.tt|" \
                      r"qr\.ae|adf\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|q\.gs|is\.gd|" \
                      r"po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|x\.co|" \
                      r"prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|" \
                      r"tr\.im|link\.zip\.net"
```

```
# 8. Checking for Shortening Services in URL (Tiny_URL)
def tinyURL(url):
    match=re.search(shortening_services,url)
    if match:
        return 1
    else:
        return 0
```

```
# 9.Checking for Prefix or Suffix Separated by (-) in the Domain (Prefix/Suffix)
def prefixSuffix(url):
    if '-' in urlparse(url).netloc:
        return 1            # phishing
    else:
        return 0            # legitimate
```

```
def has_client_in_string(url):
  if 'client' in url.lower():
    return 1
  else:
    return 0
```

```
def has_admin_in_string(url):
  if 'admin' in url.lower():
    return 1
  else:
    return 0
```

```
def has_server_in_string(url):
  if 'server' in url.lower():
    return 1
  else:
    return 0
```

```
def has_login_in_string(url):
  if 'login' in url.lower():
    return 1
  else:
    return 0
```

```
def have_php(url):
  if '.php' in url.lower():
    return 1
  else:
    return 0
```

```
def have_html(url):
  if '.html' in url.lower():
    return 1
  else:
    if '.htm' in url.lower():
      return 1
    else:
      return 0
```

```
def have_info(url):
  if '.info' in url.lower():
    return 1
  else:
    return 0
```

```
def have_text(url):
  if '.txt' in url.lower():
    return 1
  else:
    return 0
```

```
def have_js(url):
  if '.js' in url.lower():
    return 1
  else:
    return 0
```

```
def have_exe(url):
  if '.exe' in url.lower():
    return 1
  else:
    if 'install' in url.lower():
      return 1
    else:
      return 0
```

```
def number_of_periods(url):
  periods = [i for i in url if i == '.']
  return len(periods)
```

```
def is_encoded(url):
  if '%' in url.lower():
    return 1
  else:
    return 0
```

```
def num_encoded_char(url):
  encs = [i for i in url if i == '%']
  return len(encs)
```

```
def number_of_parameters(url):
  params = urlparse(url).query
  if params == '':
    return 0 
  else:
    return len(params.split('&'))
```

```
def number_of_digits(url):
  digits = [i for i in url if i.isdigit()]
  return len(digits)
```

```
# Number of Special characters

def number_spec_char(url):
  count = 0
  sp1 = [i for i in url if i == '@']
  count = count + len(sp1)
  sp2 = [i for i in url if i == '#']
  count = count + len(sp2)
  sp3 = [i for i in url if i == '$']
  count = count + len(sp3)
  sp4 = [i for i in url if i == '%']
  count = count + len(sp4)
  sp5 = [i for i in url if i == '^']
  count = count + len(sp5)
  sp6 = [i for i in url if i == '&']
  count = count + len(sp6)
  sp7 = [i for i in url if i == '*']
  count = count + len(sp7)
  sp8 = [i for i in url if i == '_']
  count = count + len(sp8)
  sp9 = [i for i in url if i == '-']
  count = count + len(sp9)
  sp10 = [i for i in url if i == '=']
  count = count + len(sp10)
  sp11 = [i for i in url if i == '+']
  count = count + len(sp11)
  sp12 = [i for i in url if i == '!']
  count = count + len(sp12)
  return count
```

Many features can be extracted that come under this category. Out of them, below mentioned were considered for this project.

*   IFrame Redirection
*   Status Bar Customization
*   Disabling Right Click
*   Website Forwarding

Each of these features are explained and the coded below:

```
import requests
from pyquery import PyQuery
import re
```

```
# 1. IFrame Redirection (iFrame)
def iframe(response):
  if response == "":
      return -1
  else:
      if re.findall(r"[<iframe>|<frameBorder>]", response.text):
          return 1
      else:
          return 0
```

```
# 2.Checks the effect of mouse over on status bar (Mouse_Over)
def mouseOver(response): 
  if response == "" :
    return -1
  else:
    if re.findall("<script>.+onmouseover.+</script>", response.text):
      return 1
    else:
      return 0
```

```
# 3.Checks the status of the right click attribute (Right_Click)
def rightClick(response):
  if response == "":
    return -1
  else:
    if re.findall(r"event.button ?== ?2", response.text):
      return 1
    else:
      return 0
```

```
# 4.Checks the number of forwardings (Web_Forwards)    
def forwarding(response):
  if response == "":
    return -1
  else:
    if len(response.history) <= 2:
      return 1
    else:
      return 0
```

```
# 5. Total number of words on page separated by ' ' excluding tags.
def number_of_page_tokens(response):
  if response == "":
    return -1
  else:
    html_tokens = len(response.text.lower().split())
    return html_tokens
```

```
# 6. Toal number of sentences on page.
def number_of_sentences(response):
  if response == "":
    return -1
  else:
    html_sentences = len(response.text.split('.'))
    return html_sentences
```

```
# 7. Total number of HTML tags.
def number_of_html_tags(response):
  try:
    if response == "":
      return -1
    else:
      pq = PyQuery(response.text)
      return len(pq('*')) if pq else None
  except:
    return -1
```

```
# 8. Total number of whitespaces.
def number_of_whitespace(response):
  if response == "":
    return -1
  else:
    whitespaces = [i for i in response if i == ' ']
    return len(whitespaces)
```

```
# 9. Check that URL is Live or not
def urlIsLive(url):
  try:
    response = requests.get(url)
    if response.ok is True:
      return 1
    else:
      return 0
  except:
    return -1
```

```
# 10. Total number of characters in HTML page excluding tags.
def HTML_Length(response):
  try:
    if response == "" :
      return -1
    else:
      pq = PyQuery(response.text)
      if pq is not None:
        return len(pq('html').text())
      else:
        return 0
  except:
    return -1
```

```
# 11. Total number of object tags.
def numObjects(response):
  try:
    if response == "":
      return -1
    else:
      pq = PyQuery(response.text)
      if pq is not None:
        return len([i for i in pq('object').items()])
      else:
        return 0
  except:
    return -1
```

```
# 12. Total number of embed tags.
def numEmbeds(response):
  try:
    if response == "":
      return -1
    else:
      pq = PyQuery(response.text)
      if pq is not None:
        return len([i for i in pq('embed').items()])
      else:
        return 0
  except:
    return -1
```

```
# 13. Total number of images embedded in HTML page.

def numImages(response):
  try:
    if response == "":
      return -1
    else:
      pq = PyQuery(response.text)
      if pq is not None:
        return len([i for i in pq('img').items()])
      else:
        return 0
  except:
    return -1
```

```
# 14. Total number of links embedded in HTML page.
def numLinks(response):
  try:
    if response == "":
      return -1
    else:
      pq = PyQuery(response.text)
      if pq is not None:
        return len([i for i in pq('a').items()])
      else:
        return 0
  except:
    return -1
```

```
# 15. Total number of H1-H6 titles in HTML page.
def numTitles(response):
  try:
    if response == "":
      return -1
    else:
      pq = PyQuery(response.text)
      if pq is not None:
        titles = ['h{}'.format(i) for i in range(7)]
        titles = [pq(i).items() for i in titles]
        return len([item for s in titles for item in s])
      else:
        return 0
  except:
    return -1
```

```
# 16. Total number of characters in embedded script in HTML page
def numScript(response):
  try:
    if response == "":
      return -1
    else:
      pq = PyQuery(response.text)
      if pq is not None:
        return len(pq('script').text())
      else:
        return 0
  except:
    return -1
```

```
# 17. Total number of special characters in HTML page
def specialCharacters(response):
  try:
    if response == "":
      return -1
    else:
      pq = PyQuery(response.text)
      if pq is not None:
        bodyText = pq('html').text()
        schars = [i for i in bodyText if not i.isdigit() and not i.isalpha()]
        return len(schars)
      else:
        return 0
  except:
    return -1
```

```
# 18. Ratio of total length of embedded script to special characters in HTML page

def scriptToSpecialCharsRatio(response):
  try:
    if response == "":
      return -1
    else:
      v = specialCharacters(response)
      pq = PyQuery(response.text)
      if pq is not None and v!=0:
        sscr = numScript(response)/v
      else:
        sscr = 0
      return sscr
  except:
    return -1
```

```
# 19. Ratio of total length of embedded script to total number of character in HTML page

def scriptTobodyRatio(response):
  try:
    if response == "" :
      return -1
    else:
      pq = PyQuery(response.text)
      v = HTML_Length(response)
      if pq is not None and v!=0:
          sbr = numScript(response)/v
      else:
          sbr = 0
      return sbr
  except:
    return -1
```

```
# 20. Ratio of total number of character to special characters in HTML page

def bodyToSpecialCharRatio(response):
  try:
    if response == "":
      return -1
    else:
      pq = PyQuery(response.text)
      v = HTML_Length(response)
      if pq is not None and v!=0:
        bscr = specialCharacters(response)/v
      else:
        bscr = 0
      return bscr
  except:
    return -1
```

## **4. Computing URL Features**

Create a list and a function that calls the other functions and stores all the features of the URL in the list. We will extract the features of each URL and append to this list.

```
#Function to extract features
def featureExtraction(url,label):

  features = []
  #Address bar based features (10)
  features.append(getDomain(url))
  features.append(havingIP(url))
  features.append(haveAtSign(url))
  features.append(getLength(url))
  features.append(getDepth(url))
  features.append(redirection(url))
  features.append(httpDomain(url))
  features.append(tinyURL(url))
  features.append(prefixSuffix(url))
  features.append(has_client_in_string(url))
  features.append(has_admin_in_string(url))
  features.append(has_login_in_string(url))
  features.append(has_server_in_string(url))
  features.append(have_php(url))
  features.append(have_html(url))
  features.append(have_info(url))
  features.append(have_text(url))
  features.append(have_js(url))
  features.append(have_exe(url))
  features.append(number_of_periods(url))
  features.append(is_encoded(url))
  features.append(num_encoded_char(url))
  features.append(number_of_parameters(url))
  features.append(number_of_digits(url))
  features.append(number_spec_char(url))
  
  #HTML & Javascript based features (4)
  try:
    response = requests.get(url)
  except:
    response = ""
    
  features.append(iframe(response))
  features.append(mouseOver(response))
  features.append(rightClick(response))
  features.append(forwarding(response))
  features.append(number_of_page_tokens(response))
  features.append(number_of_sentences(response))
  features.append(number_of_html_tags(response))
  features.append(number_of_whitespace(response))
  
  features.append(urlIsLive(url))
  features.append(HTML_Length(response))
  features.append(numObjects(response))
  features.append(numEmbeds(response))
  features.append(numImages(response))
  features.append(numLinks(response))
  features.append(numTitles(response))
  features.append(numScript(response))
  features.append(specialCharacters(response))
  features.append(scriptToSpecialCharsRatio(response))
  features.append(scriptTobodyRatio(response))
  features.append(bodyToSpecialCharRatio(response))

  features.append(label)
  
  return features
```
